```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('fivethirtyeight')
from pylab import rcParams  #для регулировки столбчатой диаграммы
```

```python
data = pd.read_csv('/home/user/Work/creditcard.csv', index_col=0, parse_dates=True)
#bmi_df = pd.read_excel(url) для экселя такая функция есть

x = data["Class"].value_counts()
#print(x)
rcParams["figure.figsize"] = 9, 12
#x.plot(kind="barh")
data.head(10)
```

```python
plt.plot(x)
plt.yscale(value="log")
```

```python
fraud_transaction = data.loc[data['Class']==1, 'V1']
normal_transaction= data.loc[data['Class']==0, 'V1']

plt.hist([fraud_transaction, normal_transaction], bins=20, density=True, color=['red','grey'], alpha=0.5)
plt.xlabel('V1')
plt.legend(labels=["Class0", "Class1"])
```
