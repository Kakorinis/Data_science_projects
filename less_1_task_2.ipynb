Задание 1 Импортируйте библиотеку Pandas и дайте ей псевдоним pd. Создайте датафрейм authors со столбцами author_id и author_name, в которых соответственно содержатся данные: [1, 2, 3] и ['Тургенев', 'Чехов', 'Островский']. Затем создайте датафрейм book cо столбцами author_id, book_title и price, в которых соответственно содержатся данные:
[1, 1, 1, 2, 2, 3, 3], ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'], [450, 300, 350, 500, 450, 370, 290]. 

```python
import pandas as pd
```

```python
a = {
    "author_id": [1, 2, 3],
    "author_name": ["Turgenev", "Chekhov", "Ostrovsky"]}

authors = pd.DataFrame(a)

b = {
    "author_id": [1, 1, 1,2,2,3,3],
    "book_title": ['Отцы и дети', 'Рудин', 'Дворянское гнездо', 'Толстый и тонкий', 'Дама с собачкой', 'Гроза', 'Таланты и поклонники'],
    "price": [450, 300, 350, 500, 450, 370, 290]}

books = pd.DataFrame(b)

```

Задание 2 Получите датафрейм authors_price, соединив датафреймы authors и books по полю author_id.

```python
authors_price = pd.merge(authors, books, on=('author_id'))
authors_price
```

Задание 3 Создайте датафрейм top5, в котором содержатся строки из authors_price с пятью самыми дорогими книгами.

```python
top5 = pd.DataFrame(authors_price.nlargest(5, "price"))
top5
```

Задание 4
Создайте датафрейм authors_stat на основе информации из authors_price. В датафрейме authors_stat должны быть четыре столбца:
author_name, min_price, max_price и mean_price,
в которых должны содержаться соответственно имя автора, минимальная, максимальная и средняя цена на книги этого автора.


```python
# способ № 1
authors_statt = authors_price.groupby("author_name").agg({'price': ['max', 'min', 'mean']})

authors_statt

# способ №2
authors_stat = pd.DataFrame({
    'author_name': [],
    'min_price': [],
    'max_price': [],
    'mean_price': []
})
# авторс-стат - словарь с массивами в виде ключей

# заполняю ключ Имена авторов:
authors_stat['author_name'] = authors_price["author_name"].unique()

# для заполнения макс, мин, ср значений цены книг по каждому автору сначала делаю словари.
# такой способ выбран поскольку др. идей по способу добавления не было
max_dict = dict(authors_price.groupby("author_name")['price'].max())
min_dict = dict(authors_price.groupby("author_name")['price'].min())
mean_dict = dict(authors_price.groupby("author_name")['price'].mean())


    
# реализовываю способ добавления цен авторов в авторс-стат, последовательность заполнения массивов соответствует


for i in range(0, len(authors_stat["author_name"])):
    name = authors_stat["author_name"][i]
    authors_stat['max_price'][i] = max_dict[name]
    authors_stat['min_price'][i] = min_dict[name]
    authors_stat['mean_price'][i] = f'{mean_dict[name]: .2f}'

    

print(authors_stat)
```
